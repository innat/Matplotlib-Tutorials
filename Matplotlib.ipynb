```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Functional Ploting

```python
'''
Functional Ploting
'''
mat_x = np.arange(10).reshape(2,5)

plt.plot(mat_x, mat_x) 
plt.xlabel('x label') # label along with X axes
plt.ylabel('y label') # label along with Y axes
plt.title('header');  # Title at top
plt.legend()
```

```python
'''
Multiple plotting at a time with functional method.
'''

x = np.linspace(1,10,30) # create 30 values with evenly spaced
y = np.exp(x)

plt.subplot(1,2,1)                      # subplot(row_number, cols_number, plot_ID)
plt.plot(x,y, 'r-', linewidth = 2)      # ploting 
plt.xlabel('First | x label')           # label along with X axes
plt.ylabel('First | y label')           # label along with Y axes
plt.title('First | Title')              # Title at top

plt.subplot(1,2,2)                      # subplot(row_number, cols_number, plot_ID)
plt.plot(y,x, 'g', linewidth = .9)      # ploting
plt.xlabel('Second | x label')          # label along with X axes
plt.ylabel('Second | y label')          # label along with Y axes
plt.title('Second | Title')             # Title at top

plt.tight_layout()                      # It helps to fit properly all the plots nicely.
```

## Object Oriented Method

```python
fig = plt.figure()                 # created a figure object 
axes = fig.add_axes([1,1,1,1])     # create axes: left-right, up-down, height, width

axes.plot(x,y, label = 'blue');    # plotting
axes.set_xlabel('X Label')         # x labeling
axes.set_ylabel('Y Label')         # y labeling
axes.set_title('Title')            # title
axes.legend();
```

```python
'''
Multiple plotting at a time with object oriented method.
'''

fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x,y)
ax[0].set_xlabel('X Label')     # x labeling
ax[0].set_ylabel('Y Label')     # y labeling
ax[0].set_title('Title');       # title

ax[1].plot(y,x)
ax[1].set_xlabel('X Label')     # x labeling
ax[1].set_ylabel('Y Label')     # y labeling
ax[1].set_title('Title');       # title
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax[0,0].plot(x,y)
ax[0,0].set_xlabel('X Label')     # x labeling
ax[0,0].set_ylabel('Y Label')     # y labeling
ax[0,0].set_title('Title');       # title

ax[0,1].plot(x,y)
ax[0,1].set_xlabel('X Label')     # x labeling
ax[0,1].set_ylabel('Y Label')     # y labeling
ax[0,1].set_title('Title');       # title


ax[1,0].plot(y,x)
ax[1,0].set_xlabel('X Label')     # x labeling
ax[1,0].set_ylabel('Y Label')     # y labeling
ax[1,0].set_title('Title');       # title

ax[1,1].plot(y,x)
ax[1,1].set_xlabel('X Label')     # x labeling
ax[1,1].set_ylabel('Y Label')     # y labeling
ax[1,1].set_title('Title');       # title
plt.tight_layout()
```

## Figure Size and Saving Plot

```python
fig , ax = plt.subplots(nrows = 2, ncols = 1, figsize = (5,5))

ax[0].plot(x,y, linewidth = 2, color = 'purple')
ax[0].set_xlabel('X Label')                        # x labeling
ax[0].set_ylabel('Y Label')                        # y labeling
ax[0].set_title('Title');                          # title

ax[1].plot(y,x,'c', lw = 2, alpha = 0.5)           # lw : shorthand of linewidth
ax[1].set_xlabel('X Label')                        # x labeling
ax[1].set_ylabel('Y Label')                        # y labeling
ax[1].set_title('Title');                          # title

plt.tight_layout()
plt.savefig('plot.png', dpi = 1000)
```

## Multiple Plots

```python
fig , ax = plt.subplots(nrows = 2, ncols = 1, figsize = (8,8))

ax[0].plot(-x,y, label = 'blue', linestyle = '-.', marker = 'o', markersize = 5)
ax[0].plot(-x,x, label = 'org')
ax[0].set_xlabel('X Label')                        # x labeling
ax[0].set_ylabel('Y Label')                        # y labeling
ax[0].set_title('Title');                          # title
ax[0].legend(loc = 'best')

ax[1].plot(y, x, linewidth = .5, color = 'purple', marker = '+', markersize = 6)
ax[1].plot(y, y, 'r', ls = '-.')                   # ls : shorthand of linestyle
ax[1].set_xlabel('X Label')                        # x labeling
ax[1].set_ylabel('Y Label')                        # y labeling
ax[1].set_title('Title');                          # title

plt.tight_layout()
```
